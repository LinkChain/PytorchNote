# 数据可视化

```python
from pandas import read_csv
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import numpy as np
filename = '../data/pima_data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = read_csv(filename, names=names)
```

```python
#直方图，分布情况：高斯分布还是指数分布还是偏态分布
data.hist(xlabelsize=10,ylabelsize=10,figsize=(10,10))
plt.show()
```

```python
#直方图，分布情况：高斯分布还是指数分布还是偏态分布
data.hist(xlabelsize=10,ylabelsize=10,figsize=(10,10), orientation='horizontal')
plt.show()
```

```python
# kde画密度图
data.plot(kind='kde',figsize=(30,10))
plt.show()
```

```python
#密度线
data.plot(kind='density', subplots=True, layout=(10,3),figsize=(10,20), sharex=False)
plt.show()
```

```python
#箱线
data.plot(kind='box', subplots=True, layout=(3,3),figsize=(10,10), sharex=False)
plt.show()
```

```python
#相关矩阵图，各个属性两两关联关系
correlations = data.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0, 9, 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

```python
#散点矩阵图
scatter_matrix(data)
plt.show()
```
